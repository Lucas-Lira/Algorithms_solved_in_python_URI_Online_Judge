```python
try:
    
    c = input().upper()

    m = []

    for line in range(12):
        m.append([])
        for column in range(12):
            m[line].append(float(input()))
           
    first_line = 1
    line = first_line
    column = 11
    last_line = 11

    if c == 'S':

        soma = 0

        while column > 6:

            while line < last_line:
                soma += m[line][column]
                line += 1

            first_line += 1
            line = first_line
            last_line -= 1
            column -= 1

        print('%.1f' % soma)

    elif c == 'M':
        
        soma = 0
        amount_of_values = 0

        while column > 6:

            while line < last_line:
                soma += m[line][column]
                amount_of_values += 1
                line += 1

            first_line += 1
            line = first_line
            last_line -= 1
            column -= 1

        media = soma/float(amount_of_values)
            
        print('%.1f' % media)
    
except:
    print('Invalid value')
```
